```python
import torch
from l2cs import Pipeline, render

gaze_pipeline = Pipeline(
    weights="/kaggle/working/Gaze360/L2CSNet_gaze360.pkl",
    arch='ResNet50',
    device=torch.device(0) # or 'gpu'
)
```

```python
import numpy as np
from PIL import Image

image = Image.open("/kaggle/working/test.jpg")
image = image.resize((512, 512))
arr = np.array(image)
image
```

```python
results = gaze_pipeline.step(arr)
gaze = render(arr, results)
Image.fromarray(gaze)
```
